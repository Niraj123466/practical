```python
df["pickup_datetime"]=pd.to_datetime(df["pickup_datetime"])
df.shape
```

```python
df=df[df["pickup_longitude"].between(-180,180)&
      df["pickup_latitude"].between(-90,90)&
      df["dropoff_longitude"].between(-180,180)&
      df["dropoff_latitude"].between(-90,90)]
df.shape
```

```python
df=df[df["fare_amount"].notnull()&(df["fare_amount"]>0)]
df.shape
```

```python
def haversine(lat1,lon1,lat2,lon2):
    R=6371
    lat1,lon1,lat2,lon2=map(np.radians,[lat1,lon1,lat2,lon2])
    a=np.sin((lat2-lat1)/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin((lon2-lon1)/2)**2
    return 2*R*np.arcsin(np.sqrt(a))
df["distance_km"]=haversine(df["pickup_latitude"],df["pickup_longitude"],
                            df["dropoff_latitude"],df["dropoff_longitude"])
    
```

```python
df=df[df["distance_km"]>0 & (df["distance_km"]<200)]
df.shape
```

```python
df.isnull().sum()
```

```python
columns=['fare_amount','distance_km']
df.boxplot(columns)
```

```python
def drop_outlier(data,col):
    q1,q3=data[col].quantile([0.25,0.75])
    iqr=q3-q1
    low,high=q1-1.5*iqr,q3+1.5*iqr
    return data[(data[col]>=low)&(data[col]<=high)]

for c in columns:
    df=drop_outlier(df,c)
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
df['hour'] = df['pickup_datetime'].dt.hour
df['dayofweek'] = df['pickup_datetime'].dt.dayofweek
df['month'] = df['pickup_datetime'].dt.month
```

```python
df
```

```python
sns.heatmap(df[["fare_amount","hour","dayofweek","distance_km","month"]].corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
x=df[["distance_km","hour","dayofweek","month"]]
y=df["fare_amount"]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
xtrain_scaled=scaler.fit_transform(xtrain)
xtest_scaled=scaler.transform(xtest)
```

```python
linear=LinearRegression()
linear = LinearRegression()
linear.fit(xtrain_scaled, ytrain)
ypred=linear.predict(xtest_scaled)
print("Linear:")

print("R2_Score:",r2_score(ytest,ypred))
print("RMSE:",np.sqrt(mean_squared_error(ytest,ypred)))
```

```python
ytrain_pred = linear.predict(xtrain_scaled)
ytest_pred = linear.predict(xtest_scaled)
print("R2_train:",r2_score(ytrain, ytrain_pred))
print("R2_test:", r2_score(ytest, ytest_pred))
print("RMSE_train:", np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
print("RMSE_test:",np.sqrt(mean_squared_error(ytest, ytest_pred,)))
print("MAE_train:", mean_absolute_error(ytrain, ytrain_pred))
print("MAE_test:", mean_absolute_error(ytest, ytest_pred))
```
